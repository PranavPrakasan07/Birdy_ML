# Model Building

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Convolution2D, MaxPooling2D, Flatten
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale = 1/255, shear_range = 0.2, zoom_range = 0.2, horizontal_flip = True)
```

```python
test_datagen = ImageDataGenerator(rescale = 1/255)
```

```python
x_train = train_datagen.flow_from_directory('../input/100-bird-species/train', target_size = (64,64), batch_size = 32, class_mode = 'categorical')
```

```python
x_test = test_datagen.flow_from_directory('../input/100-bird-species/test', target_size = (64,64), batch_size = 32, class_mode = 'categorical')
```

```python
x_train.class_indices
```

```python
model = Sequential()
```

```python
model.add(Convolution2D(32, (3,3), input_shape = (64, 64, 3))) # 3 indicated coloured image
```

```python
model.add(MaxPooling2D((2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units = 128, kernel_initializer = 'random_uniform', activation = 'relu'))
```

```python
model.add(Dense(units = 300, kernel_initializer = 'random_uniform', activation = 'relu'))
```

```python
model.add(Dense(units = 300, kernel_initializer = 'random_uniform', activation = 'relu'))
```

```python
model.add(Dense(units = 300, kernel_initializer = 'random_uniform', activation = 'softmax'))
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit_generator(x_train, steps_per_epoch = 1363, epochs = 100, validation_data = x_test, validation_steps = 46)
```

```python
model.save('birdModel.h5')
```
